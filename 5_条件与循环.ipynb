# 条件与循环
## 1. 认识
1. if 不加括号
2. python中的数据结构只要是可迭代的（iterable）, 可以通过for-in遍历
3. 遍历中使用index，需要使用range + len 函数 / enumerate
4. while

## 2. 注意点
1. 空字符串、0、None、空集合 都是False
2. 字典只有key是可遍历的, 可以通过d.values()遍历值、items 遍历开k、v




```
# 一个判断语句 y = |x|
x = 4
if x >= 0:
  y = x
else:
  y = -x
print(y)
```

```
# elif
a = 1
if a == 1:
  print('1-route')
elif a == 2:
  print("2-route")
else:
  print('not support')
```

```
# 循环语句
l = [1, 3, 4, 5, 5]
for item in l:
  print(item)
```

```
# 字典遍历出来的是key, 可以通过key再范围value
d = {'name': 'zhangsan', 'age': 11}
for k in d:
  print(k)
  print(d.get(k))

print('----------')

# .values() 遍历值
for v in d.values():
  print(v)

# .items 遍历 k、v
for k,v in d.items():
  print('name:{}, age:{}'.format(k, v))
```

```
# 遍历list 基于 index

l = ['a', 'b', 'c', 'd', 'e']
for index in range(0, len(l)):
  print('index:{}, value:{}'.format(index, l[index]))

print('----------------')

# 使用enumerate更简洁
for index, item in enumerate(l):
  print("index:{}, value:{}".format(index, item))
```

```
# while 使用实例
l = ['a', 'b', 'c', 'd', 'e', 'f']
i = 0
while i < 3:
  print('item:{}'.format(l[i]))
  i = i + 1
```

```
# continue的使用
l = ['a', 'b', 'c', 'd', 'e', 'f']
for index, item in enumerate(l):
  if(index == 0):
    continue
  print('index:{}, value:{}'.format(index, item))
```

```
# break 使用
l = ['a', 'b', 'c', 'd', 'e', 'f']
for index, item in enumerate(l):
  if(index == 3):
    break
  print('index:{}, value:{}', index, item)
```
