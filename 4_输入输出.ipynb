# 输入输出
## 1. 认识
1. 命令行输入输出：input 函数， 暂停函数运行，等待用户输入，直到用户按回车，参数是提示语言
2. 文件输入输出: open 、r/w 、read()
3. json序列化: json.dumps json.loads

## 2. 注意点
1.open函数如果没有使用with, 要主动close()



```
# 命令行输入输出
name = input("请输入用户名")
print(name)
```

```
# 直接读文件
# open 函数拿到文件的句柄
# r是读权限，w表示写, rw表示读写
# read() 会直接读取文件所有内容，缺点文件如果过大，会内存崩溃，可以指定size，或者使用readline按行读取
with open('test.txt', 'r') as fin :
  text = fin.read()
print(text)
```

```
# 使用json 序列化
import json

d = {
    'name': 'zangsan',
    'age': 11
}

# 序列化
json_str = json.dumps(d)
print(type(json_str))
print(json_str)

print('--------------')

# 反序列化
d2 = json.loads(json_str)
print(type(d2))
print(d2)
```
