# 匿名函数
## 1. 认识
1. python 支持lambda，通过匿名函数实现
2. 格式：lambda argument1, argument2,... argumentN : expression
3. 参数可以为空 lambda: expression

4. python 不是函数式编程语言，但是提供了一些函数编程特性


## 2. 注意点
1. lambda的主体是只有一行的简单表达式，并不能扩展成一个多行的代码块
2. 注意在python3中map函数返回的是一个 惰性的 map 对象（迭代器），不是列表。一旦把 map 转成 list 取完之后，再用同一个 map 对象就不会再有元素了（因为迭代器只能消费一次）
3. python3中reduce不再是内置函数，被移到functools


```
# lambda 简单示例

add = lambda a, b: a + b
ret = add(1, 2)
print("ret:{}".format(ret))
print(add)
```

```
# 函数式编程特性 - map(function, iterable)
# 对iterable中的每个元素，都运用function这个函数，最后返回一个新的可遍历的集合
# 在python3中map返回的是一个懒惰的map对象，不是列表，需要转一下

l = [1, 2, 3, 4, 5]
new_list = map(lambda x: x * 2, l)
print(list(new_list))
```

```
# 函数式编程特性 - filter(function, iterable)

filter_ret_list = filter(lambda x: x % 2 == 0, [1,2,3,4,5])
print(list(filter_ret_list))
```

```
# 函数式编程特性 - reduce(function, iterable)
# iterable中的每个元素以及上一次调用后的结果,执行function操作
# 注意首次，没有上一次结果，所以直接从列表取两个结果作为参数
# 最后返回一个结果
from functools import reduce

reduce_ret = reduce(lambda a,b: a + b, [1, 2, 3, 4, 5])
print(reduce_ret)
```
