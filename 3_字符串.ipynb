# 字符串
## 1. 认识
1. 单引号（''）双引号（""）或者三引号之中（''' '''或""" """，两者一样）
2. 字符串本质是一个字符元组，可以遍历，但是不可以修改

## 2. 注意点
1. 使用'''时，开始的''' 要和变量在一行



```
# 定义字符串可以用 ' '' ''' 都一样
s1 = 'abc'
print(s1)

# 用于内嵌带引号的字符串
s2 = "I'm zhangsan"
print(s2)

# 用于多行字符串,
s3 = ''' name: I'm zhangsan
age: 12
'''
print(s3)

# 注意如果'''不合变量定义在一行，就会报错
s3 =
'''
name: I'm zhangsan
age: 12
'''
```

```
# 遍历字符串
s = 'abcde'
for c in s:
  print(c)
```

```
# 通过index修改字符串中的字符，会报错
s = 'abcde'
print(s[0])
s[0] = 'm'
```

```
# 使用replace修改字符串
s = 'aabcd'
print(s)
# 注意replace后会返回一个新的字符串，原字符串不会修改
s.replace('a', 'A')
print(s)
print(s.replace('a', 'A'))
```

```
# 通过 + 修改字符串
s = 'aabcd'
print(s)
s = s + 'ef'
print(s)
```

```
# 字符串切片
s = 'aabcd'
print(s[1:])
print(s[1:3])
```
