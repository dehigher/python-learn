# 异常处理
## 1. 认识
1. 分类：语法错误、异常
- 语法错误：不符合Python的语法规范，所以程序就会报错invalid syntax
- 当程序运行到这些地方时，就抛出了异常，并且终止运行(退出)， XXXError
2. try-except 捕获异常，注意except 后面可以不接异常，except: 相当于 except Exception as err: , 但是忽略了Exception就拿不到error info, 不建议使用

3. Exception 顶级异常

4. 用户自定义异常 class MyExceptioin(Exception), 后面展开

5. 常见异常：
- ValueError
- IndexError
- ZeroDivisionError
- NameError
- TypeError

## 2. 注意点
1. 当程序中存在多个except block时，最多只有一个except block会被执行。换句话说，如果多个except声明的异常类型都与实际相匹配，那么只有最前面的except block会被执行，其他则被忽略
2. 不同except block 生命的err变量名可以重复




```
# 捕获指定异常
try:
  10/0
except ZeroDivisionError as err:
  print('Error info: {}'.format(err))
```

```
# 捕获指定异常并兜底
try:
  10/0
except IndexError as err:
  print('IndexError info: {}'.format(err))
except Exception as err:
  print('Exception info: {}'.format(err))
```
