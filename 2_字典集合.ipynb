# 字典&集合
## 1. 认识
1. 字典其实就是java 里的map, pythoy里叫 dict，集合其实就是set
2. pythoy3.7之后dict是有序的，之前无法保证, 集合无序
3. dict的比较，单key、value个数、字面值相同时为true, 否则false

## 2. 注意点
1. json格式定义， key也要加引号，可以用单引号, 也可以用双引号
2. dict 有多种定义方式，最简单推荐的就是用json


```
# 定义 、使用 字典
d1 = {"name": "zhangsan", "age": 11, 'a': 0}
print(d1)
d2 = dict({'name': 'zhangsan', 'age': 11})
print(d2)
d3 = dict([('name', 'zhangsan'), ('age', 11)])
print(d3)
d4 = dict(name = 'zhangsan', age = 11)
print(d4)
print(d1 == d2 == d3 == d4)
print(d2 == d3 == d4)

# 取值
print(d1.get('name'))
print(d1.get('nnnn', 'dv'))
```

```
# 定义使用集合
s = {'hello', 1, 2}
print(s)
```

```
# 判读key是否在字典中
d = {'name': 'zhangsan', 'age': 1}
print( 'age' in d)
```

```
# 判断是否在集合中
s = {1, 2, 3}
print( 4 in s )
```

```
# 字典基本操作
d = {'name': 'zhangsan', 'age': 11}
print(d)
print(d['name'])
print(d.get('name'))
# 新增k-v
d['gender'] = 'male'
print(d)
# 由于3.7之后dist是有序的，后加入的在后面，可以可以使用pop指定key, 返回value
print(d.pop('age'))
print(d)
# 添加相同key, 直接覆盖value
d['name'] = 'lisi'
print(d)
```

```
# 集合基本操作
s = {1, 2, 3, 4, 5}
print(s)
# 移出指定元素
s.remove(3)
print(s)
# 添加
s.add(6)
print(s)
# 添加相同元素, 添加不进去
s.add(1)
print(s)
```
