# 自定义函数
## 1. 认识
1. def 关键字
2. python函数支持嵌套，嵌套的作用内部函数只能被外部函数访问（权限控制）
3. python默认函数内访问的变量都是局部变脸。如果要访问全局变量，只能读不能写，写的话会报错，需要用global声明下。
4. 类似对于嵌套函数，内部函数范围外部函数变量只能读不能改，需要改的话用nonlocal声明，注意不会报错，只改副本，不改原值

5. python 支持闭包

## 2. 注意点
1. def 语句是懒执行的，只有函数被调用时，才会去创建一个函数对象
2. python 的函数参数可以使用默认值
3. python 函数参数不约束类型, 返回值也不约束，无需声明类型

```
# 一个简单的函数定义&函数调用

#定义
def say_message(message):
  print('say:{}'.format(message))

#调用
say_message('hello world')
```

```
# python 函数可以设置默认值
def add(a = 1, b = 1):
  return a + b

# 主动赋值
print(add(2, 2))

# 使用默认值
print(add())

# 使用部分默认值
print(add(2))


# 也可以给参数穿字符串,只要函数逻辑支持
print(add('abc', 'ef'))


```

```
# python 函数嵌套，内部函数只能被外部函数使用

def add(a , b):
  print('a + b = {}'.format(a + b))
  def mul(a, b):
    print('a * b = {}'.format(a * b))
  mul(a, b)

# 调用
add(11, 22)

# 在外部调用mul
mul(11, 22)
```

```
# 直接访问全局变量 - 只读
MAX_VALUE = 10
l = [1, 4, 5 , 7, 13, 15]

# 输出小于10的数
def printNumLess10(l):
  for item in l:
    if item <= MAX_VALUE:
      print("item:{}".format(item))
    else:
      continue;

# 调用
printNumLess10(l)
```

```
# 直接访问全局变量 - 写
MAX_VALUE = 10
l = [1, 4, 5 , 7, 13, 15]

# 输出小于10的数
def printNumLess10(l):
  for item in l:
    if item <= MAX_VALUE:
      print("item:{}".format(item))
    else:
      MAX_VALUE = item

# 调用
printNumLess10(l)
```

```
# 直接访问全局变量 - 写
MAX_VALUE = 10
l = [1, 4, 5 , 7, 13, 15]

# 输出小于10的数
def printNumLess10(l):
  global MAX_VALUE
  for item in l:
    if item <= MAX_VALUE:
      print("item:{}".format(item))
    else:
      MAX_VALUE = item

# 调用
printNumLess10(l)
print(MAX_VALUE)
```

```
# 嵌套函数-内部函数修改外部函数的变量, 没有用nolocal声明，不会报错，但是修改的是副本，不影响外部变量

def outter_fun():
    x = "1"
    def inner_fun():
      x = "2"
      print("inner_x:{}".format(x))
    inner_fun()
    print("outter_x:{}".format(x))

# 调用
outter_fun()
```

```
# 嵌套函数-内部函数修改外部函数的变量, 用nonlocal声明，可以改动原值

def outter_fun():
    x = "1"
    def inner_fun():
      nonlocal x
      x = "2"
      print("inner_x:{}".format(x))
    inner_fun()
    print("outter_x:{}".format(x))

# 调用
outter_fun()
```

```
# 闭包 closure
# 一个函数引用外部环境

de = 8

# 计算一个函数的n次方
def nth_util(n):
  def exponent_of(base):
    return base ** n + de
  return exponent_of

ntu = nth_util(2)

ret = ntu(3)

print(ret)


```
