## 2. 注意点

1. 私有属性，用self.__xxx 表示 xxx 私有属性
2. 规范：全大写来表示常量， 可以在类中使用 self.WELCOME_STR ，或者在类外使用 Entity.WELCOME_STR ，来表达这个字符串
3. 继承类在生成对象的时候，是不会自动调用父类的构造函数的，因此你必须在 init()函数中显式调用父类的构造函数

```
# 一个类的定义
class Document():
  # 构造函数
  def __init__(self, title, author, context):
    print('init function called')
    self.title = title
    self.author = author
    self.__context = context

  def get_context_lenth(self):
    return len(self.__context)

  def intercept_context(self, length):
    return self.__context[:length]




# 调用构造函数创建对象
document = Document('Harry Potter', 'J. K. Rowling', '... Forever Do not believe any thing is capable of thinking independently ...')
print("title:{}\n author:{},\n ".format(document.title, document.author))

# 私有属性，只能用内部方法访问
print(document.intercept_context(10))
```

```
# 声明和使用类变量

class Document():
  # 定义类变量
  WELCOME_STR = "Welcome! The context for this book is {}."


  def __init__(self, title, author, context):
    self.title = title
    self.author = author
    self.__context = context



  def getWelcomeStr(self):
    # 使用类变量，注意要使用类名引用
    return Document.WELCOME_STR.format(self.title)


# 使用
document1 = Document('Harry Potter', 'J. K. Rowling', '... Forever Do not believe any thing is capable of thinking independently ...')
print(document1.getWelcomeStr())
```

```
# 类函数、成员函数、静态函数
# 静态函数说明：和当前类、和对象都没关系，没有 cls参数，也没有self参数
# 注意静态方法也通过实例调用

class Document():
  # 定义类变量
  WELCOME_STR = "Welcome! The context for this book is {}."


  def __init__(self, title, author, context):
    self.title = title
    self.author = author
    self.__context = context


  # 成员函数
  def getContextInfo(self):
    return self.__context

  # 类函数
  @classmethod
  def buildEmptyBook(cls, title):
    return cls(title, "system", "nothing")


  # 静态函数
  @staticmethod
  def sayhello(user):
    print("{}, hello!!".format(user))



document1 = Document('Harry Potter', 'J. K. Rowling', '... Forever Do not believe any thing is capable of thinking independently ...')
print(document1.getContextInfo())

print('---------')

document2 = Document.buildEmptyBook('class build book')
print(document2.getContextInfo())

print("---------")
document2.sayhello("zhangsan")


```

```
# 继承
class Entity():
  def __init__(self, object_type):
    print('parent class init called')
    self.object_type = object_type

  def get_context_length(self):
    raise Exception('not support')

  def print_title(self):
    print(self.title)

# Document 继承 Entity
class Document(Entity):

  def __init__(self, title, author, context):
    print('Document class init called')
    Entity.__init__(self, 'document')
    self.title = title
    self.author = author
    self.__context = context

  # 重写父类方法
  def get_context_length(self):
    return len(self.__context)


doc1 = Document('title1', 'author1', 'context:000000000000000')
print("length:{}".format(doc1.get_context_length()))

print('--------')
doc1.print_title()

```

```
# 抽象类、抽象方法 - 抽象类不能直接实例化
from abc import ABCMeta, abstractmethod

class Entity(metaclass = ABCMeta):

  @abstractmethod
  def get_title(self):
    pass

  @abstractmethod
  def set_title(self):
    pass

# 实现抽象类（init方法可以不写）
class Document(Entity):

  def get_title(self):
    return self.title

  def set_title(self, title):
    self.title = title

document = Document()
document.set_title('title...')
print(document.get_title())
```
