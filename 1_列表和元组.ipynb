# 列表和元组
1. 列表:动态的 元组tuple:静态的 列表list用[]、元组用()
2. 在python中集合中的元素不要求类型一样, 但是不建议，sort等内置方法报错
3. 元组不光不能改长度，也不能改元素内容
4. pythoy列表和原组都支持负索引，-1 表示最后一个元素，  -2表示倒数第二个
5. 列表和元组都支持切片，下标区间含前不含后
6. 列表和元组的转换非常简单


特别注意
1. 不要再使用 list、tuple 这些内置类型名作为变量名

```
# 定义一个元组， 类型可以不一致
list = [1,2, "hello", "world"]
list
```

```
# 定义一个元组
arr = ('json', 22)
arr
```

```
# 根据下标访问列表元素, 下标从0开始
list = ["a", "b", "c"]
print(list[1])
# 根据下标改元素
list[1] = 'm'
print(list[1])
```

```
arr = ('a', 'b', 'c', 'd')
print(arr[0])
# 根据下标修改元素报错
arr[0] = 'n'
```

```
# 通过新建元组的方式实现元组修改
old_arr = (1, 2, 3, 4, 5)
print(old_arr)
new_arr = old_arr + (6, 7)
print(new_arr)
```

```
# 通过负索引访问元组
arr = ('a', 'b', 'c', 'd', 'e')
print(arr[-1])
print(arr[-2])
```

```
# 元组和列表都支持切片,下标区间含前不含后
arr = ('a', 'b', 'c', 'd', 'e', 'f')
print(arr[1:4])
```

```
# 列表、元组相互转换
my_list = [1, 2, 3]
arr = tuple(my_list)
print(arr)
print(list(arr))
```

```
# 数组和元组常见的内置函数
l = [3,4,4,5,5,5,7,8,0]
# 1. 计算指定元素个数
print(l.count(4))
# 2. 查指定元素的位置, 返回首次出现的位置
print(l.index(4))
# 3. reverse
print(l)
l.reverse()
print(l)
# 4.sort, 默认按照自然顺序排，存在不同元素时报错
# l.sort()
print(sorted(l))

```

```
# 列表新增\移除元素，当传入元素存在多个时，只移除index最小的一个
l = [1, 3,2, 3, 3]
l.append(5)
print(l)
l.remove(3)
print(l)
```
